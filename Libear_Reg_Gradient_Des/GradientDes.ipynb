```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/content/drive/MyDrive/ML/LinearRegression/house_data.csv')
```

```python
df.shape
```

```python
x = df["SquareFeet"]
y = df["Price"]
```

```python
plt.scatter(x, y)
plt.xlabel("SquareFeet")
plt.ylabel("Price")
plt.show()
```

```python
x_scaled = (x - np.mean(x)) / np.std(x)
y_scaled = (y - np.mean(y)) / np.std(y)
```

```python
plt.scatter(x_scaled, y_scaled)
plt.xlabel("SquareFeet")
plt.ylabel("Price")
plt.show()
```

# Gradient desent in ML

```python
# Step 5: Gradient Descent Initialization
m = 0
c = 0
l = 0.01      # learning rate
iterations = 500
n = len(x_scaled)
```

```python
# Step 6: Gradient Descent Loop
cost_history = []

for i in range(iterations):
    y_pred = m * x_scaled + c
    cost = (1/n) * np.sum((y_scaled - y_pred) ** 2)
    cost_history.append(cost)

    derivatives_m = (-2/n) * np.sum(x_scaled * (y_scaled - y_pred))
    derivatives_c = (-2/n) * np.sum(y_scaled - y_pred)

    m = m - l * derivatives_m
    c = c - l * derivatives_c
```

```python
# Step 7: Convert slope & intercept back to original scale
m_original = m * (np.std(y) / np.std(x))
c_original = c * np.std(y) + np.mean(y)

print("Slope (m):", m_original)
print("Intercept (c):", c_original)
```

```python
plt.plot(range(iterations), cost_history)
plt.xlabel("Iterations")
plt.ylabel("Cost (MSE)")
plt.title("Gradient Descent Cost Reduction")
plt.show()
```
